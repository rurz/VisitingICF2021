```julia
using VisitingICF2021
using PyPlot
```

# 1. Rectangular modes

Call the function $\Phi_{2}(n_{x}, n_{y}, j)$ who gives the two-dimensional Kravchuk functions of the oscillator.

```julia
function plΦ2(n₁, n₂, j)
    v₀ = zeros(Float64, (2 * j + 1, 2 * j + 1))
    for q₁ in -j:j
        for q₂ in -j:j
            v₀[q₁ + j + 1, q₂ + j + 1] = Φ2(n₁, n₂, q₁, q₂, j)
        end
    end
    return v₀
end
;
```

Reinterpretation $\Phi_{2}(n, m, j)$ in terms of the total energy mode $n = n_{x} + n_{y}$  and the counter $m = n_{x} - n_{y}$

```julia
nmΦ2(m, n, j) = plΦ2((1/2) * (m + n), abs((1/2) * (m - n)), j);
```

Selected combination of modes

```julia
rm1 = nmΦ2(1, 1, 16)
rm32 = nmΦ2(32, 32, 16)
rm63 = nmΦ2(63, 1, 16)
;
```

Plot the modes

```julia
fig_01, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (7, 5))

ax1.imshow(rm1, cmap = "gray")
    ax1.set_title("(n, m) = (1, 1)")
    ax1.set_xticks([])
    ax1.set_yticks([])
ax2.imshow(rm32, cmap = "gray")
    ax2.set_title("(n, m) = (32, 32)")
    ax2.set_xticks([])
    ax2.set_yticks([])
ax3.imshow(rm63, cmap = "gray")
    ax3.set_title("(n, m) = (63, 1)")
    ax3.set_xticks([])
    ax3.set_yticks([])

fig_01.suptitle("Energy modes of the 2D Kravchuk Oscillator", x = 0.5, y = 0.8)

tight_layout()
```

# 2. Rotated rectangular modes

Call the function $\Phi_{2}^{(\pi/2)}(n, m, j)$ who gives the two-dimensional rotated energy modes of the Kravchuk oscillator. They are mediated by the aplication of the little-$d$ function $d_{n,m}^{j}(\pi/2)$

```julia
function plRΦ(n, m, j)
    v₁ = zeros((2 * j + 1, 2 * j + 1))
    for q₁ in -j:j
        for q₂ in  -j:j
            if 0 <= n <= 2 * j
                v₁[q₁ + j + 1, q₂ + j + 1] = VisitingICF2021.Φd(q₁, q₂, n, m, j)
            elseif 2 * j < n <= 4 * j
                v₁[q₁ + j + 1, q₂ + j + 1] = VisitingICF2021.Φu(q₁, q₂, n, m, j)
            end
        end
    end
    return v₁
end;
```

Selected combitation of modes

```julia
Rm1 = plRΦ(1, 1, 16)
Rm32 = plRΦ(32, 32,16)
Rm63 = plRΦ(63, -1,16)
;
```

Plot the modes

```julia
fig_02, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (7, 5))

ax1.imshow(Rm1, cmap = "gray")
    ax1.set_title("(n, m) = (1, 1)")
    ax1.set_xticks([])
    ax1.set_yticks([])
ax2.imshow(Rm32, cmap = "gray")
    ax2.set_title("(n, m) = (32, 32)")
    ax2.set_xticks([])
    ax2.set_yticks([])
ax3.imshow(Rm63, cmap = "gray")
    ax3.set_title("(n, m) = (63, 1)")
    ax3.set_xticks([])
    ax3.set_yticks([])

fig_02.suptitle("Rotated energy modes of the 2D Kravchuk Oscillator", x = 0.5, y = 0.8)

tight_layout()
```
